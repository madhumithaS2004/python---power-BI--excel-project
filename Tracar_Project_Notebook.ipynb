```python
import folium, pandas, time, random, threading
from datetime import datetime
from IPython.display import display, clear_output
try:
    import ipywidgets as widgets
    WIDGETS_AVAILABLE = True
except:
    WIDGETS_AVAILABLE = False

class TracarRecorder:
    def __init__(self, start_lat=12.9716, start_lon=77.5946):
        self.history=[]
        self.current={'lat':start_lat,'lon':start_lon}
        self.running=False
    def add_location(self,lat,lon,note=''):
        ts=datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S')
        self.history.append({'lat':lat,'lon':lon,'ts':ts,'note':note})
    def get_map(self):
        if not self.history:
            return folium.Map(location=[self.current['lat'],self.current['lon']],zoom_start=14)
        m=folium.Map(location=[self.history[-1]['lat'],self.history[-1]['lon']],zoom_start=14)
        pts=[(h['lat'],h['lon']) for h in self.history]
        for h in self.history:
            folium.CircleMarker([h['lat'],h['lon']],radius=4,popup=h['ts']).add_to(m)
        folium.PolyLine(pts,color='blue').add_to(m)
        return m
recorder=TracarRecorder()
```

```python
recorder.add_location(12.9716,77.5946,'Start Point')
display(recorder.get_map())
```

```python
import time, random
for i in range(5):
    lat = recorder.history[-1]['lat'] + random.uniform(-0.001, 0.001)
    lon = recorder.history[-1]['lon'] + random.uniform(-0.001, 0.001)
    recorder.add_location(lat, lon, f"Step {i+1}")
    display(recorder.get_map())
    time.sleep(1)
    clear_output(wait=True)
```
